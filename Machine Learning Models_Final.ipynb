```python
#Initially loading few Libraries 

import os
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
df["Total Scenes"]= df["Average Positive Sentiment"] + df["Average Negative Sentiment"] + df["Average Neutral Sentiment"]
```

```python
df["Positive Sentiment_Score"]= df["Average Positive Sentiment"]/ df["Total Scenes"]
```

```python
df["Negative Sentiment_Score"]= df["Average Negative Sentiment"]/ df["Total Scenes"]
```

```python
df["Neutral Sentiment_Score"]= df["Average Neutral Sentiment"]/ df["Total Scenes"]
```

```python
df["Anger_prop"]= df["Anger"]/ df["Total Scenes"]
df["Anticipation_prop"]= df["Anticipation"]/ df["Total Scenes"]
df["Disgust_prop"]= df["Disgust"]/ df["Total Scenes"]
df["Fear_prop"]= df["Fear"]/ df["Total Scenes"]
df["Joy_prop"]= df["Joy"]/ df["Total Scenes"]
df["Sadness_prop"]= df["Sadness"]/ df["Total Scenes"]
df["Surprise_prop"]= df["Surprise"]/ df["Total Scenes"]
df["Trust_prop"]= df["Trust"]/ df["Total Scenes"]

```

```python
df
```

```python
df['log(ROI+1)'].describe()
```

```python
df['Movie Rating'].describe()
```

```python
df.drop(columns='score',inplace=True)
```

```python
df['std_log']=df['log(ROI+1)'].apply(lambda x: (x-np.mean(df['log(ROI+1)']))/np.std(df['log(ROI+1)']))
```

```python
df['std_Movie_Rating']=df['Movie Rating'].apply(lambda x: (x-np.mean(df['Movie Rating']))/np.std(df['Movie Rating']))
```

```python
df['score'] = 0.6*df['std_log'] + 0.4*df['std_Movie_Rating']
```

```python
from sklearn.preprocessing import MinMaxScaler
df['std_score'] = MinMaxScaler().fit_transform(np.array(df['score']).reshape(-1,1))*10
```

```python
from sklearn.neighbors import KNeighborsRegressor

test_scores = []
train_scores = []

for i in range(1,25):

    knn = KNeighborsRegressor(i)
    knn.fit(X_train,y_train)
    
    train_scores.append(knn.score(X_train,y_train))
    test_scores.append(knn.score(x_val,Y_val))
```

```python
## score that comes from testing on the datapoints that were split in the beginning to be used for testing solely
max_test_score = max(test_scores)
test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
print('Max test score {} % and k = {}'.format(max_test_score*100,list(map(lambda x: x+1, test_scores_ind))))
```
